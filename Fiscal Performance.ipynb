<p> <b>Methodology</b>

<b>To ensure robust and reliable insights, a rigorous data processing pipeline was implemented:</b>

<b>Data Cleaning:</b> A total of 23,784 observations were carefully examined. Mixed data types in the Amount column were resolved by removing commas, parsing numeric values, and converting the Time variable into standardized datetime format.

<b>Unit Standardization:</b> Significant heterogeneity existed across measurement units (Millions vs. Billions). All financial variables were therefore standardized to Millions of local currency, ensuring internal consistency within each country.

<b>Currency Handling:</b> Given country-specific currencies (e.g., EGP, NGN, ZAR, etc), cross-country aggregation was avoided to prevent misleading comparisons. Analysis instead emphasized within-country dynamics and normalized indicators such as fiscal deficits as a percentage of GDP.

<b>Aggregation:</b> High-frequency monthly flow variables, including budget deficits and revenues, were aggregated into annual totals to ensure alignment with yearly macroeconomic indicators such as GDP.

```python
anomly.dropna(inplace = True)
anomly.rename(columns ={'Budget Deficit/Surplus':'deficit'}, inplace =True)
#anomly.loc[2025] = {"deficit": -13080165}
#anomly = anomly.sort_index()
anomly.head(2)
```

```python
anomly.index = pd.to_numeric(anomly.index, errors="coerce")
anomly = anomly[anomly.index.notna()].sort_index()

# Replace with your actual deficit column name
deficit_col = "deficit"

deficit = pd.to_numeric(anomly[deficit_col], errors="coerce")
# ANOMALY DETECTION (Z-SCORE)
mean = deficit.mean()
std = deficit.std()

z_score = (deficit - mean) / std
anomalies = deficit[np.abs(z_score) > 2]   # threshold = 2

# PLOTTING
fig, ax = plt.subplots(figsize=(12, 6))
# Line plot
ax.plot(
    deficit.index,
    deficit,
    marker="o",
    linewidth=3,
    color ='#0f5452',
    label="Deficit")
# Highlight anomalies
ax.scatter(
    anomalies.index,
    anomalies,
    color="red",
    s=120,
    label="Anomaly",
    zorder=5)
ax.set_title("Anomaly Detection in Nigeria Budget Deficit")
ax.set_xlabel("Year")
ax.set_ylabel("Amount (NGN Millions)")

ax.legend()
ax.grid(True, alpha=0.3)

plt.savefig(r"C:\Users\hp\Desktop\Projects\Python\Fiscal Performance Project\12.png",dpi=300, bbox_inches="tight")
plt.tight_layout()
plt.show()
```

```python
# Ensure Year index is integer
anomly.index = pd.to_numeric(anomly.index, errors="coerce").astype(int)
anomly = anomly.sort_index()

deficit = pd.to_numeric(anomly["deficit"], errors="coerce")


# FORECAST SETUP
# Use last N years to estimate trend
N = 5
x_hist = deficit.index[-N:]
y_hist = deficit.values[-N:]

# Linear trend: y = mx + c
coef = np.polyfit(x_hist, y_hist, 1)
trend = np.poly1d(coef)

# Forecast years
forecast_years = np.arange(x_hist.max() + 1, x_hist.max() + 4)
forecast_values = trend(forecast_years)


# PLOTTING
fig, ax = plt.subplots(figsize=(12, 6))

# Historical data
ax.plot(
    deficit.index,
    deficit/ 1e7,
    color="#0f5452",
    linewidth=3.5,
    label="History"
)

# Forecast
ax.plot(
    forecast_years,
    forecast_values/ 1e7,
    linestyle="--",
    marker="o",
    color="red",
    linewidth=3.5,
    label="Forecast"
)

# CLEAN FORMATTING
ax.set_facecolor("white")
fig.patch.set_facecolor("white")

# Integer years on x-axis
ax.set_xticks(
    list(deficit.index) + list(forecast_years)
)
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x)}"))

# Y-axis in trillions
ax.yaxis.set_major_formatter(
    FuncFormatter(lambda y, _: f"{y:.1f}")
)

ax.set_ylabel("Amount (₦ Mrillion)")
ax.set_xlabel("")
ax.set_title("Nigeria Fiscal Forecast (2026–2028)", fontsize =14 )

ax.grid(True, alpha=0.7)
ax.legend()

plt.savefig(r"C:\Users\hp\Desktop\Projects\Python\Fiscal Performance Project\13.png",dpi=300, bbox_inches="tight")
plt.tight_layout()
plt.show()
```

This reallocation erodes fiscal space, weakens long-term human capital development, and constrains inclusive growth.

Against this backdrop, the following policy recommendations are proposed;

-<b>First</b>, the government should pursue gradual expenditure rationalization by eliminating non-essential and inefficient spending. While an abrupt reduction in the cost of governance may be impractical, a phased approach—such as incremental cuts to administrative and operational expenditures—can steadily reduce fiscal leakages without disrupting public service delivery.

-<b>Second</b>, enhanced financial intelligence and prioritization, akin to credit rationing in financial markets, should guide public investment decisions. The government must rigorously assess which capital and social projects yield the highest economic and social returns, while postponing 
or discontinuing projects with limited impact. This targeted allocation of resources would improve spending efficiency and fiscal sustainability.

-<b>Finally</b>, expanding and diversifying revenue streams must take precedence over continued expenditure expansion. Nigeria’s heavy reliance on oil revenues exposes public finances to external shocks, particularly crude oil price volatility. Strengthening revenue mobilization from non-oil sectors—such as manufacturing, agriculture, and services—offers a more resilient and sustainable fiscal base, enhancing the government’s capacity to finance development priorities.</p>

```python
# Ensure Year index is integer
ano.index = pd.to_numeric(ano.index, errors="coerce").astype(int)
ano = ano.sort_index()

deficit = pd.to_numeric(ano["deficit"], errors="coerce")


# FORECAST SETUP
# Use last N years to estimate trend
N = 5
x_hist = deficit.index[-N:]
y_hist = deficit.values[-N:]

# Linear trend: y = mx + c
coef = np.polyfit(x_hist, y_hist, 1)
trend = np.poly1d(coef)

# Forecast years
forecast_years = np.arange(x_hist.max() + 1, x_hist.max() + 4)
forecast_values = trend(forecast_years)


# PLOTTING
fig, ax = plt.subplots(figsize=(12, 6))

# Historical data
ax.plot(
    deficit.index,
    deficit/ 1e7,
    color="#0f5452",
    linewidth=3.5,
    label="History"
)

# Forecast
ax.plot(
    forecast_years,
    forecast_values/ 1e7,
    linestyle="--",
    marker="o",
    color="red",
    linewidth=3.5,
    label="Forecast"
)

# CLEAN FORMATTING
ax.set_facecolor("white")
fig.patch.set_facecolor("white")

# Integer years on x-axis
ax.set_xticks(
    list(deficit.index) + list(forecast_years)
)
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x)}"))

# Y-axis in trillions
ax.yaxis.set_major_formatter(
    FuncFormatter(lambda y, _: f"{y:.1f}")
)

ax.set_ylabel("Amount (₦ Mrillion)")
ax.set_xlabel("")
ax.set_title("Fiscal Trajectory Forecast (2026–2028)", fontsize =14 )

ax.grid(True, alpha=0.7)
ax.legend()

plt.tight_layout()
plt.show()
```
